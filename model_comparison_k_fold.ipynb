```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

models = [
    "MobileNetV2",
    "InceptionV3",
    "ResNet50",
    "Swin Transformer",
    "ViT",
    "EfficientNet-B0",
    "DenseNet121",
    "Proposed CXRNet",
]

fold1 = [88.02, 90.12, 91.88, 93.21, 93.52, 93.98, 94.12, 98.97]
fold2 = [88.76, 90.95, 92.64, 93.88, 94.02, 94.41, 94.65, 99.17]
fold3 = [88.51, 90.91, 92.59, 93.89, 94.10, 94.18, 94.55, 99.43]
means = [88.43, 90.66, 92.37, 93.66, 93.88, 94.19, 94.44, 99.19]
stds  = [0.38,  0.46,  0.42,  0.39,  0.31,  0.22,  0.28,  0.23]

n = len(models)
x = np.arange(n)
width = 0.2

BASE_COLORS = {
    "fold1": "#B5D4F4",
    "fold2": "#9FE1CB",
    "fold3": "#F0997B",
    "mean":  "#AFA9EC",
}
HIGHLIGHT_COLORS = {
    "fold1": "#378ADD",
    "fold2": "#1D9E75",
    "fold3": "#D85A30",
    "mean":  "#7F77DD",
}

def bar_colors(key):
    return [HIGHLIGHT_COLORS[key] if i == n - 1 else BASE_COLORS[key] for i in range(n)]

fig, axes = plt.subplots(1, 2, figsize=(16, 7), gridspec_kw={"width_ratios": [2, 1]})
fig.patch.set_facecolor("white")

ax = axes[0]
ax.set_facecolor("white")

b1 = ax.bar(x - 1.5 * width, fold1, width, color=bar_colors("fold1"), label="Fold 1", zorder=2)
b2 = ax.bar(x - 0.5 * width, fold2, width, color=bar_colors("fold2"), label="Fold 2", zorder=2)
b3 = ax.bar(x + 0.5 * width, fold3, width, color=bar_colors("fold3"), label="Fold 3", zorder=2)
b4 = ax.bar(x + 1.5 * width, means, width, color=bar_colors("mean"),  label="Mean",   zorder=2)

ax.set_ylim(85, 101)
ax.set_xticks(x)
ax.set_xticklabels(models, rotation=25, ha="right", fontsize=10)
ax.set_ylabel("Accuracy (%)", fontsize=11)
ax.set_title("Cross-Validation Accuracy by Model", fontsize=13, fontweight="bold", pad=12)
ax.yaxis.grid(True, linestyle="--", alpha=0.5, zorder=0)
ax.set_axisbelow(True)
ax.spines[["top", "right"]].set_visible(False)

legend_patches = [
    mpatches.Patch(color=HIGHLIGHT_COLORS["fold1"], label="Fold 1"),
    mpatches.Patch(color=HIGHLIGHT_COLORS["fold2"], label="Fold 2"),
    mpatches.Patch(color=HIGHLIGHT_COLORS["fold3"], label="Fold 3"),
    mpatches.Patch(color=HIGHLIGHT_COLORS["mean"],  label="Mean"),
]
ax.legend(handles=legend_patches, fontsize=10, framealpha=0.6, loc="lower right")

ax2 = axes[1]
ax2.set_facecolor("white")

colors_mean = [HIGHLIGHT_COLORS["mean"] if i == n - 1 else BASE_COLORS["mean"] for i in range(n)]
bars = ax2.barh(x, means, xerr=stds, color=colors_mean,
                error_kw=dict(ecolor="#555", capsize=4, lw=1.2), zorder=2)
ax2.set_xlim(85, 101)
ax2.set_yticks(x)
ax2.set_yticklabels(models, fontsize=10)
ax2.set_xlabel("Mean Accuracy (%)", fontsize=11)
ax2.set_title("Mean ± Std Accuracy", fontsize=13, fontweight="bold", pad=12)
ax2.xaxis.grid(True, linestyle="--", alpha=0.5, zorder=0)
ax2.set_axisbelow(True)
ax2.spines[["top", "right"]].set_visible(False)

for bar, mean, std in zip(bars, means, stds):
    ax2.text(
        bar.get_width() + std + 0.15,
        bar.get_y() + bar.get_height() / 2,
        f"{mean:.2f}",
        va="center", ha="left", fontsize=9,
        color="#333",
    )

fig.suptitle(
    "CXRNet vs Baseline Models — 3-Fold Cross Validation",
    fontsize=14, fontweight="bold", y=1.01
)
plt.tight_layout()
plt.savefig("model_comparison.png", dpi=150, bbox_inches="tight", facecolor="white")
plt.show()
print("Saved: model_comparison.png")
```
