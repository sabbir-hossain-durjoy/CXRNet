```python
import os
import numpy as np
import matplotlib.pyplot as plt
# =====================================================
# CREATE OUTPUT FOLDER (KAGGLE)
# =====================================================
output_dir = "/kaggle/working/"
os.makedirs(output_dir, exist_ok=True)
# =====================================================
# GLOBAL WHITE BACKGROUND SETTINGS
# =====================================================
plt.style.use('default')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['savefig.edgecolor'] = 'white'
# =====================================================
# Example Loss Values
# =====================================================
# -------------------------
# Fold 1
# -------------------------
fold1_train_loss = [
0.45,0.18,0.09,0.04,0.03,0.02,0.018,0.012,0.010,0.009,
0.011,0.013,0.014,0.012,0.011,0.012,0.010,0.011,0.009,0.010,
0.009,0.011,0.010,0.011,0.010,0.013,0.015,0.012,0.011,0.012
]
fold1_val_loss = [
0.15,0.08,0.05,0.04,0.035,0.037,0.032,0.030,0.031,0.029,
0.030,0.031,0.032,0.034,0.028,0.030,0.033,0.034,0.032,0.035,
0.036,0.035,0.034,0.033,0.032,0.040,0.045,0.034,0.031,0.033
]
# -------------------------
# Fold 2
# -------------------------
fold2_train_loss = [
0.48,0.17,0.08,0.05,0.03,0.02,0.021,0.013,0.012,0.013,
0.011,0.010,0.011,0.012,0.009,0.008,0.010,0.011,0.010,0.009,
0.011,0.012,0.010,0.009,0.008,0.012,0.010,0.011,0.015,0.010
]
fold2_val_loss = [
0.17,0.09,0.04,0.04,0.045,0.028,0.050,0.041,0.039,0.032,
0.031,0.032,0.030,0.031,0.033,0.029,0.041,0.033,0.031,0.034,
0.035,0.034,0.036,0.035,0.039,0.043,0.040,0.041,0.027,0.045
]
# -------------------------
# Fold 3
# -------------------------
fold3_train_loss = [
0.50,0.16,0.07,0.04,0.025,0.020,0.021,0.022,0.009,0.012,
0.011,0.013,0.010,0.010,0.007,0.008,0.012,0.011,0.010,0.009,
0.008,0.009,0.008,0.009,0.010,0.011,0.010,0.012,0.013,0.011
]
fold3_val_loss = [
0.18,0.07,0.05,0.04,0.035,0.038,0.036,0.037,0.030,0.032,
0.031,0.030,0.031,0.028,0.027,0.026,0.033,0.034,0.029,0.060,
0.040,0.031,0.050,0.033,0.034,0.031,0.037,0.036,0.034,0.035
]
# =====================================================
# Epochs
# =====================================================
epochs = np.arange(1, 31)
# =====================================================
# Figure 1: Loss Curves Across All Folds
# =====================================================
fig, ax = plt.subplots(figsize=(14,8), facecolor='white')
ax.set_facecolor('white')
# NO GRID
ax.grid(False)
# Fold 1
ax.plot(
    epochs,
    fold1_train_loss,
    linewidth=2.5,
    label='Fold 1 Train Loss'
)
ax.plot(
    epochs,
    fold1_val_loss,
    linestyle='--',
    linewidth=2.5,
    label='Fold 1 Validation Loss'
)
# Fold 2
ax.plot(
    epochs,
    fold2_train_loss,
    linewidth=2.5,
    label='Fold 2 Train Loss'
)
ax.plot(
    epochs,
    fold2_val_loss,
    linestyle='--',
    linewidth=2.5,
    label='Fold 2 Validation Loss'
)
# Fold 3
ax.plot(
    epochs,
    fold3_train_loss,
    linewidth=2.5,
    label='Fold 3 Train Loss'
)
ax.plot(
    epochs,
    fold3_val_loss,
    linestyle='--',
    linewidth=2.5,
    label='Fold 3 Validation Loss'
)
# Labels and title
ax.set_xlabel('Epoch', fontsize=16)
ax.set_ylabel('Loss', fontsize=16)
# Axis settings
ax.set_xlim(1, 30)
# Legend
ax.legend(
    fontsize=10,
    loc='upper right',
    frameon=True
)
# Layout
plt.tight_layout()
# Save figure
plt.savefig(
    os.path.join(output_dir, 'KFold_Loss_Comparison.png'),
    dpi=600,
    bbox_inches='tight',
    facecolor='white'
)
# Show
plt.show()
# =====================================================
# Figure 2: Mean Validation Loss ± STD
# =====================================================
val_loss_matrix = np.array([
    fold1_val_loss,
    fold2_val_loss,
    fold3_val_loss
])
mean_val_loss = np.mean(val_loss_matrix, axis=0)
std_val_loss = np.std(val_loss_matrix, axis=0)
fig, ax = plt.subplots(figsize=(12,7), facecolor='white')
ax.set_facecolor('white')
# NO GRID
ax.grid(False)
# Mean validation loss
ax.plot(
    epochs,
    mean_val_loss,
    linewidth=3,
    label='Mean Validation Loss'
)
# Standard deviation area
ax.fill_between(
    epochs,
    mean_val_loss - std_val_loss,
    mean_val_loss + std_val_loss,
    alpha=0.20,
    label='Standard Deviation'
)
# Labels and title
ax.set_xlabel('Epoch', fontsize=16)
ax.set_ylabel('Validation Loss', fontsize=16)
# Legend
ax.legend(
    fontsize=12,
    loc='upper right',
    frameon=True
)
# Layout
plt.tight_layout()
# Save figure
plt.savefig(
    os.path.join(output_dir, 'Mean_Validation_Loss_STD.png'),
    dpi=600,
    bbox_inches='tight',
    facecolor='white'
)
# Show
plt.show()
# =====================================================
# Print Summary
# =====================================================
print("="*60)
print("Loss Curve Summary")
print("="*60)
print(f"Fold 1 Minimum Validation Loss: {min(fold1_val_loss):.4f}")
print(f"Fold 2 Minimum Validation Loss: {min(fold2_val_loss):.4f}")
print(f"Fold 3 Minimum Validation Loss: {min(fold3_val_loss):.4f}")
print("-"*60)
print(
    f"Mean Minimum Validation Loss: "
    f"{np.mean([min(fold1_val_loss), min(fold2_val_loss), min(fold3_val_loss)]):.4f}"
)
print("="*60)
print(f"\nFigures saved in: {output_dir}")
```
