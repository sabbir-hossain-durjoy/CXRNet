```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams

rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Arial', 'Helvetica', 'DejaVu Sans']
rcParams['font.size'] = 10
rcParams['axes.titlesize'] = 12
rcParams['axes.labelsize'] = 11
rcParams['xtick.labelsize'] = 9
rcParams['ytick.labelsize'] = 9
rcParams['legend.fontsize'] = 9

models = [
    'DenseNet121', 'EfficientNet-B0', 'InceptionV3', 'MobileNetV2',
    'ResNet50', 'Swin Transformer', 'Vision Transformer (ViT)',
    'CXRNet(Raw)', 'CXRNet(Augmented)'
]

accuracies = np.array([94.44, 94.19, 90.66, 88.43, 92.37, 93.66, 93.88, 98.57, 99.56])

misclassifications = 100 - accuracies

x = np.arange(len(models))

fig, ax = plt.subplots(figsize=(8.0, 4.0))

ax.axhline(100, linestyle='--', linewidth=1.2, label='Perfect Classification (100%)')

for i, acc in enumerate(accuracies):
    ax.plot([x[i], x[i]], [acc, 100], linewidth=1, color='red', alpha=0.7)

ax.scatter(x, accuracies, s=40, zorder=3, label='Model Accuracy')

for i, misclass in enumerate(misclassifications):
    mid_y = (accuracies[i] + 100) / 2
    ax.annotate(f'{misclass:.2f}%', xy=(x[i], mid_y), xytext=(0.15, 0),
                textcoords='offset points', fontsize=8, ha='left', va='center',
                bbox=dict(boxstyle='round,pad=0.2', facecolor='white', alpha=0.8),
                arrowprops=dict(arrowstyle='->', color='gray', lw=0.5))

ax.set_title('Dot–Line Deviation Plot of Classification Performance',
             fontweight='bold', pad=10)
ax.set_xlabel('Models', fontweight='bold')
ax.set_ylabel('Accuracy (%)', fontweight='bold')

ax.set_xticks(x)
ax.set_xticklabels(models, fontsize=8, rotation=45, ha='right')

ax.set_ylim(85, 101)
ax.legend(loc='lower right', frameon=True)

ax.grid(axis='y', linestyle='--', alpha=0.4)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig('dot_line_deviation_Q1.png', dpi=300, bbox_inches='tight')
plt.show()
```
