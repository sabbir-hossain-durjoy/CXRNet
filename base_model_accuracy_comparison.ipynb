```python
import os
import matplotlib.pyplot as plt
import numpy as np

output_dir = "/kaggle/working/output"
os.makedirs(output_dir, exist_ok=True)

# Model names
models = [
    "DenseNet121",
    "EfficientNet-B0",
    "InceptionV3",
    "MobileNetV2",
    "ResNet50",
    "Swin Transformer",
    "Vision Transformer (ViT)"
]

# Accuracy values
accuracies = [94.44, 94.19, 90.66, 88.43, 92.37, 93.66, 93.88]

# Create figure
plt.figure(figsize=(10, 5))

# Bar plot
bars = plt.bar(
    models,
    accuracies,
    color='#E69F00',   # orange/golden color
    edgecolor='none',
    width=0.75
)

# Add value labels on top of bars
for bar, acc in zip(bars, accuracies):
    plt.text(
        bar.get_x() + bar.get_width()/2,
        acc + 0.2,
        f'{acc:.2f}%',
        ha='center',
        va='bottom',
        fontsize=9
    )

# Axis labels
plt.xlabel("Transfer Learning Model", fontsize=11)
plt.ylabel("Classification Accuracy (%)", fontsize=11)

# Y-axis limits
plt.ylim(80, 100)

# Rotate x-axis labels
plt.xticks(rotation=25, ha='right')

# Improve layout
plt.tight_layout()


save_path = os.path.join(output_dir, "model_accuracy_comparison.png")

plt.savefig(
    save_path,
    dpi=300,
    bbox_inches='tight'
)

print(f"Figure saved at: {save_path}")

plt.show()
```
