```python
import matplotlib.pyplot as plt
import numpy as np

models = [
    'DenseNet121', 'EfficientNet-B0', 'InceptionV3', 'MobileNetV2',
    'ResNet50', 'Swin Transformer', 'Vision Transformer (ViT)',
    'CXRNet(Raw)', 'CXRNet(Augmented)'
]

precisions = [95.05, 94.62, 89.00, 88.85, 90.84, 93.80, 93.63, 98.47, 99.54]
recalls = [95.21, 94.75, 89.22, 86.39, 92.50, 93.37, 92.82, 98.76, 99.55]
f1_scores = [95.12, 94.68, 89.08, 87.60, 91.67, 93.55, 93.20, 98.60, 99.55]
accuracies = [94.44, 94.19, 90.66, 88.43, 92.37, 93.66, 93.88, 98.57, 99.56]

plt.figure(figsize=(10, 5))

x = np.arange(len(models))

plt.plot(x, precisions, marker='o', color='red', linewidth=2, label='Avg Precision', markersize=6)
plt.plot(x, recalls, marker='s', color='blue', linewidth=2, label='Avg Recall', markersize=6)
plt.plot(x, f1_scores, marker='^', color='yellow', linewidth=2, label='Avg F1-score', markersize=6)
plt.plot(x, accuracies, marker='d', color='green', linewidth=2, label='Accuracy', markersize=6)

plt.title('Performance Metrics of Individual Models', fontsize=14, fontweight='bold')
plt.xlabel('Models', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)

plt.ylim(75, 100)

plt.xticks(x, models, rotation=45, ha='right')

plt.legend(loc='lower right')

plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()

plt.show()
```
